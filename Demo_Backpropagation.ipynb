```
import torch
import torch.nn as nn
import torch.optim as optim

# 1. PRZYGOTOWANIE DANYCH (Sztuczny przykład)
# Wejście: Liczby [1.0, 2.0, 3.0]
# Oczekiwany wynik (Cel): [2.0, 4.0, 6.0] (czyli funkcja y = 2x)
inputs = torch.tensor([[1.0], [2.0], [3.0]])
targets = torch.tensor([[2.0], [4.0], [6.0]])

# 2. BUDOWA PROSTEJ SIECI NEURONOWEJ
# Jedna warstwa liniowa - sieć ma jedną wagę (na początku losową)
model = nn.Linear(1, 1)

print(f"Startowa waga (losowa): {model.weight.item():.4f}")

# 3. DEFINICJA NARZĘDZI
criterion = nn.MSELoss()                # Funkcja straty (Jak bardzo się mylimy?)
optimizer = optim.SGD(model.parameters(), lr=0.01) # Optymalizator (Zmieniacz wag)

# --- PĘTLA UCZENIA (TRENING) ---
print("\nRozpoczynam trening...")

for epoch in range(100):
    # KROK A: FORWARD PASS (Propagacja w przód)
    # Sieć zgaduje wynik
    predictions = model(inputs)

    # Obliczamy błąd (Loss)
    loss = criterion(predictions, targets)

    # KROK B: BACKPROPAGATION (Propagacja wsteczna)
    # To jest kluczowy moment prezentacji!
    optimizer.zero_grad()   # Czyścimy stare gradienty
    loss.backward()         # <--- TU DZIEJE SIĘ MAGIA (Obliczanie pochodnych)

    # KROK C: AKTUALIZACJA WAG (Gradient Descent)
    optimizer.step()        # Przesuwamy wagi w stronę mniejszego błędu

    if (epoch+1) % 20 == 0:
        print(f"Epoka {epoch+1}: Błąd = {loss.item():.4f}, Obecna waga = {model.weight.item():.4f}")

print(f"\nWynik po treningu: Waga powinna być bliska 2.000 -> Jest: {model.weight.item():.4f}")
```
